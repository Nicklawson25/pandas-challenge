```python
import pandas as pd

# Load the datasets
schools_df = pd.read_csv('schools_complete.csv')
students_df = pd.read_csv('students_complete.csv')

# Merge the datasets on the school name
merged_df = pd.merge(students_df, schools_df, on='school_name')
merged_df.head()
```

```python
#District Summary 
```

```python
# Calculate the key metrics for the district
total_schools = schools_df['school_name'].nunique()
total_schools
```

```python
total_students = students_df['Student ID'].nunique()
total_students
```

```python
total_budget = schools_df['budget'].sum()
total_budget
```

```python
average_math_score = students_df['math_score'].mean()
average_math_score
```

```python
average_reading_score = students_df['reading_score'].mean()
average_reading_score
```

```python
passing_math = merged_df[merged_df['math_score'] >= 70].shape[0] / total_students * 100
passing_math
```

```python
passing_reading = merged_df[merged_df['reading_score'] >= 70].shape[0] / total_students * 100
passing_reading
```

```python
overall_passing = merged_df[(merged_df['math_score'] >= 70) & (merged_df['reading_score'] >= 70)].shape[0] / total_students * 100
overall_passing
```

```python
# Create a DataFrame to display the district's key metrics
district_summary = pd.DataFrame({
    'Total Schools': [total_schools],
    'Total Students': [total_students],
    'Total Budget': [total_budget],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [passing_math],
    '% Passing Reading': [passing_reading],
    '% Overall Passing': [overall_passing]
})
district_summary
```

```python
#school summary
```

```python
# Calculate the necessary metrics for each school
school_summary = merged_df.groupby(['school_name', 'type']).agg({
    'Student ID': 'count',
    'budget': 'mean',
    'math_score': 'mean',
    'reading_score': 'mean'
}).rename(columns={
    'Student ID': 'Total Students',
    'budget': 'Total School Budget',
    'math_score': 'Average Math Score',
    'reading_score': 'Average Reading Score'
})

# Calculate per student budget
school_summary['Per Student Budget'] = school_summary['Total School Budget'] / school_summary['Total Students']

# Calculate % passing math
school_summary['% Passing Math'] = merged_df[merged_df['math_score'] >= 70].groupby('school_name')['Student ID'].count() / school_summary['Total Students'] * 100

# Calculate % passing reading
school_summary['% Passing Reading'] = merged_df[merged_df['reading_score'] >= 70].groupby('school_name')['Student ID'].count() / school_summary['Total Students'] * 100

# Calculate % overall passing
school_summary['% Overall Passing'] = merged_df[(merged_df['math_score'] >= 70) & (merged_df['reading_score'] >= 70)].groupby('school_name')['Student ID'].count() / school_summary['Total Students'] * 100

school_summary.reset_index(inplace=True)

# Display the merged school summary data

school_summary
```

```python
## Highest-Performing Schools (by % Overall Passing)
```

```python
# Highest-Performing Schools (by % Overall Passing)
top_schools = school_summary.sort_values('% Overall Passing', ascending=False).head(5)
top_schools.head()
```

```python
## Bottom Performing Schools (By % Overall Passing)
```

```python
# Lowest-Performing Schools (by % Overall Passing)
bottom_schools = school_summary.sort_values('% Overall Passing').head(5)
bottom_schools.head(5)
```

```python
## Math Scores by Grade
```

```python
# Calculate average math scores by grade
grades_order = ['9th', '10th', '11th', '12th']
math_scores_by_grade = merged_df.pivot_table(values='math_score', index='school_name', columns='grade', aggfunc='mean')
math_scores_by_grade = math_scores_by_grade[grades_order]
math_scores_by_grade
```

```python
#Reading Score by Grade
```

```python
# Calculate average reading scores by grade
grades_order = ['9th', '10th', '11th', '12th']
reading_scores_by_grade = merged_df.pivot_table(values='reading_score', index='school_name', columns='grade', aggfunc='mean')
reading_scores_by_grade = reading_scores_by_grade[grades_order]
reading_scores_by_grade
```

```python
#Scores by School Spending 
```

```python
# Define spending bins and labels
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Add Spending Ranges (Per Student) to school_summary
school_summary['Spending Ranges (Per Student)'] = pd.cut(school_summary['Per Student Budget'], bins=spending_bins, labels=spending_labels)
```

```python
# Define size bins and labels
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Add Size Ranges to school_summary
school_summary['School Size'] = pd.cut(school_summary['Total Students'], bins=size_bins, labels=size_labels)
school_summary
```

```python
# Group by spending ranges and calculate the mean scores
spending_math_scores = school_summary.groupby("Spending Ranges (Per Student)")["Average Math Score"].mean()
spending_reading_scores = school_summary.groupby("Spending Ranges (Per Student)")["Average Reading Score"].mean()
spending_passing_math = school_summary.groupby("Spending Ranges (Per Student)")["% Passing Math"].mean()
spending_passing_reading = school_summary.groupby("Spending Ranges (Per Student)")["% Passing Reading"].mean()
overall_passing_spending = school_summary.groupby("Spending Ranges (Per Student)")["% Overall Passing"].mean()

# Create the spending summary DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})
spending_summary
```
